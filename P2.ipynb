```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

**1ª questão**

y’ = y com y(0) = 1 para encontrar um y(t) muito familiar

```python
def model1(y, t):
  dydt = y
  return dydt

# condição inicial
y0 = 1

# t
t = np.linspace(0, 5)

# resolvendo a ODE
y = odeint(model1, y0, t)

# solução analítica:
# y(t) = exp(t)
y1 = np.exp(t)

# plotando
plt.plot(t, y, 'b')
plt.plot(t, y1, 'r--')
plt.xlabel('tempo')
plt.ylabel('y(t)')
plt.legend(['Solução numérica', 'Solução analítica'])
plt.title('y’ = y')
plt.show()
```

**2ª questão**

y’ = 1 com y(0) = 4, uma pergunta trivial.

```python
def model2(y, t):
  dydt = 1
  return dydt

# condição inicial
y0 = 4

# t
t = np.linspace(0, 10)

# resolvendo a ODE
y = odeint(model2, y0, t)

# solução analítica:
# integrando dy/dt = 1 -> y(t) = t + c
# substituindo y(0) = 4 -> y(0) = 0 + c -> c = 4
# logo, y(t) = t + y0
y1 = t + y0

# plotando
plt.plot(t, y, 'b')
plt.plot(t, y1, 'r--')
plt.xlabel('tempo')
plt.ylabel('y(t)')
plt.legend(['Solução numérica', 'Solução analítica'])
plt.title('y’ = 1')
plt.show()
```

**3ª questão**

y’ - 6y = exp(-t), y(0) = 2;

```python
def model3(y, t):
  dydt = 6*y + np.exp(-t)
  return dydt

# condição inicial
y0 = 2

# t
t = np.linspace(0, 0.8)

# resolvendo a ODE
y = odeint(model3, y0, t)

# solução analítica:
# yp = Y*exp(-t), substituindo em y’ - 6y = exp(-t):
# -Y*exp(-t) - 6*Y*exp(-t) = exp(-t)
# -Y - 6*Y = 1 -> -7*Y = 1 -> Y = -1/7
# logo, yp = -1/7*exp(-t)
# yn = y(0)*exp(at), a = 6 -> yn = y0*exp(6t)
# y = yn + yp = y0*exp(6t) - 1/7*exp(-t)
y1 = y0*np.exp(6*t) - 1/7*np.exp(-t)


# plotando
plt.plot(t, y, 'b')
plt.plot(t, y1, 'r--')
plt.xlabel('tempo')
plt.ylabel('y(t)')
plt.legend(['Solução numérica', 'Solução analítica'])
plt.title('y’ - 6y = exp(-t)')
plt.show()
```

**4ª questão**

y’’ + 9y = 1, y(0) = y’(0) = 0.

```python
def model4(y, t):
  # y = y[0] e y' = y[1]
  # z = dydt
  # y'' + 9*y = 1 -> z + 9*y = 1 -> z = 1 - 9*y
  dydt = y[1]
  dzdt = 1 - 9*y[0]
  return [dydt, dzdt]

# condição inicial
y0 = [0, 0]

# t
t = np.linspace(0, 5)

# resolvendo a ODE
y = odeint(model4, y0, t)

# solução analítica:
# yp = k
# 9k = 1 -> k = 1/9
# yn = c1*exp(s1*t) + c2*exp(s2*t)
# s^2 + 9 = 0 -> s1 = -3i, s2 = 3i
# logo, yn = c1*exp(-3*i*t) + c2*exp(3*i*t)
# encontrando c1 e c2
# y(0) = 1/9 + c1 + c2 = 0 -> c1 + c2 = -1/9 (*)
# y'(0) = -3*i*c1 + 3*i*c2 = 0 -> c1 = c2
# substituindo em (*) -> 2c1 = -1/9 -> c1 = c2 = -1/18
# logo, y = yn + yp = 1/9 - 1/18*(exp(-3*i*t) + exp(3*i*t))
# utilizando a identidade de euler:
# exp(-3*i*t) = cos(3t) - i*sen(3t) 
# exp(3*i*t) = cos(3t) + i*sen(3t)
# logo, exp(-3*i*t) + exp(3*i*t) = 2*cos(3t)
# então teremos nossa solução como:
# y = 1/9 - cos(3t)/9
y1 = 1/9 - np.cos(3*t)/9


# plotando
plt.plot(t, y[:,0], 'b')
plt.plot(t, y1, 'r--')
plt.xlabel('tempo')
plt.ylabel('y(t)')
plt.legend(['Solução numérica', 'Solução analítica'])
plt.title('y’’ + 9y = 1')
plt.show()
```
